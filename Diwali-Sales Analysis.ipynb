Exploratory Data Analysis

Count of Male and Female

```python
ax=sns.countplot(x='Gender',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

From above graph, it is concluded that the count of Female is more than Male

Sales by Gender

```python
sales_gender = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Gender', y='Amount', data=sales_gender)
```

From above graph, it is shown that sales generated by females is more than male.

```python
ax=sns.countplot(x='Age Group',data=df,hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
From above graph, it is observed that the age group 26-35 is spending more and most of buyers are female
```

Sales by Age Group

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
ax=sns.barplot(x='Age Group', y='Amount', data=sales_age)
```

From above graph, we can observe that the age group 26-35 is spending more on shopping

Top 10 States by Order 

```python
sales_age=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(18,5)})
ax=sns.barplot(x='State',y='Orders',data=sales_age)
for bars in ax.containers:
    ax.bar_label(bars)
```

From above plot, it is observed that most of the sales is from Uttarpradesh

Top 5 States by Amount

```python
sales_states=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(5)

sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(x='State',y='Amount',data=sales_states)
```

From above graph we can observe that the amount spend by UP, Maharashtra and Karanataka are more

```python
sales_marital=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(10,5)})
ax=sns.barplot(x='Marital_Status',y='Amount',data=sales_marital,hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

From above plot it is observed that the married female are spending more amount

Sales by Occupation

```python
sales_occ=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(5)
sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(x='Occupation',y='Amount',data=sales_occ)
```

Maximum sales is from IT sector, Healthcare, Aviation 

Orders by Product Category

```python
sales_order=df.groupby(['Product_Category'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(5)
ax=sns.barplot(x='Product_Category',y='Orders',data=sales_order)

for bars in ax.containers:
    ax.bar_label(bars)
```

Clothing & Apparel, Food have highest order values

Sales by Product Category 

```python
sales_cate=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(5)
ax=sns.barplot(x='Product_Category',y='Amount',data=sales_cate)

for bars in ax.containers:
    ax.bar_label(bars)
```

Food, Clothing & Apparel, Electronics Gadgets have highest sales 

Maximum Ordered Products

```python
sales_prod=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(5)
ax=sns.barplot(x='Product_ID',y='Orders',data=sales_prod)

for bars in ax.containers:
    ax.bar_label(bars)
```

Conclusion

```python
Married Womens from 26-30 age group from UP, Maharashtra, Karnataka working in IT, Healthcare and Aviations are more likely to buy products 
from food, clothing and electronics category
```
